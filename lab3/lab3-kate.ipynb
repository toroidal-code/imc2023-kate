```python
import numpy as np
import matplotlib.pylab as plt
import librosa.display
import IPython.display
import librosa
```

```python
def plotAudio2(sig,sr,title):
    
    # this version expects the audio file to have already been opened by librosa
    
    # Adjust the x-axis to represent time
    time=np.arange(0,sig.shape[0])/sr
    
    # Scale the y-axis to represent amplitude from -1 to 1
    excursion=max(abs(sig))
    sigNew = sig / excursion
    
    # Label the x and y axes with variables
    xlabel = 'Time (seconds)'
    ylabel = 'Amplitude'

    # Plot the signal using the x and y axis variables    
    plt.plot(time,sigNew)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.show()
```

```python
def plotAudioFreqDomain(sig, sr, title, winSize, specType):
    spec = librosa.amplitude_to_db(np.abs(librosa.stft(sig, n_fft=winSize)), ref=np.max)

    # display the normalized spectrogram with a linear frequency scale
    librosa.display.specshow(spec, y_axis=specType)
    plt.colorbar(format='%+2.0f dB')
    plt.title(title)
    plt.show()
```

```python
def plotAudioFunc(sig, sr, title, winSize, specType):
    plotAudio2(sig, sr, title);
    plotAudioFreqDomain(sig, sr, title, winSize, specType)
```

```python
sig, sr = librosa.load("../audioFiles/avm.wav")
IPython.display.Audio(data=sig, rate=sr)
```

```python
plotAudioFunc(sig, sr, title='Ave Maria', winSize=2048, specType='linear')
```

```python
plotAudioFunc(sig, sr, title='Ave Maria', winSize=2048, specType='log')
```

```python
plotAudioFunc(sig, sr, title='Ave Maria', winSize=256, specType='linear')
```

```python
plotAudioFunc(sig, sr, title='Ave Maria', winSize=256, specType='log')
```
