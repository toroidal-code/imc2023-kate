## Import Libraries

```python
import numpy as np
import matplotlib.pylab as plt
import librosa
```

## Define the plotting function

```python
def plot_audio_file(filepath, sample_rate):
    """Plot the time-domain representation of a given waveform

    Args:
        filepath: path to the file containing the waveform to visualize
        sample_rate: the sample rate we want to analyze the waveform using
    Returns:
        None
    Side Effects:
        Generates a plot of the given waveform
    """
    signal, sr = librosa.load(path=filepath, sr=sample_rate)
    time = np.arange(0, signal.shape[0])/sr

    x_label = "Time (Seconds)"
    y_label = "Amplitude"

    signal_max = max(abs(signal))
    signal_scaled = signal / signal_max
    plt.plot(time, signal_scaled)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
```

## Plot with the correct sampling rate (44.1kHz)

```python
plot_audio_file("../audioFiles/avm.wav", 44100)
```

## Plot with an incorrect sampling rate (441hz)

```python
plot_audio_file("../audioFiles/avm.wav", 441)
```
